```python
import pymysql

db = pymysql.connect(host='localhost', user='root', password='mysql123')
cursor = db.cursor()

try:
    cursor.execute('create database movie_db;')
    db.commit()
    db.close()
    print('Database has been created')
    
except:
    print('Database already exsist!')
```

```python
import pandas as pd

from sqlalchemy import create_engine

db = pymysql.connect(host='127.0.0.1', user='root', password='mysql123')
cursor = db.cursor()

# csv파일 불러오기
df = pd.read_csv(r"C:\Users\HyeJin\Desktop\MultiProject\jin\movieList.csv", encoding="cp949")

# sqlalchemy를 사용해 원하는 database에 csv파일 저장
try:
    engine = create_engine("mysql+pymysql://root:mysql123@127.0.0.1/movie_db?charset=utf8mb4")
    conn = engine.connect()
    df.to_sql(name = "movielist", con = engine, index = False)
    conn.close()
    
except ValueError:
    print('Table already exists!')
```
