```python
import requests
from os import environ as env
from bs4 import BeautifulSoup
```

```python
key=env.get('e7b8d01640d7db313d377f7daac5c193')
url=f'http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieList.xml?key=e7b8d01640d7db313d377f7daac5c193'
```

```python
response = requests.get(url)
xml_data=response.content
soup=BeautifulSoup(xml_data, 'xml')
```

```python
movies = soup.find_all('movie')
```

```python
movies_list=[]
```

```python
movies_list = []

for movie in movies:
    movieCd_tag = movie.find('movieCd') #영화코드
    movieCd = movieCd_tag.text if movieCd_tag is not None else None

    movieNm_tag = movie.find('movieNm') # 영화명(국문)
    movieNm = movieNm_tag.text if movieNm_tag is not None else None

    movieNmEn_tag = movie.find('movieNmEn') # 영화명(영문)
    movieNmEn = movieNmEn_tag.text if movieNmEn_tag is not None else None

    openDt_tag = movie.find('openDt') # 개봉일
    openDt = openDt_tag.text if openDt_tag is not None else None

    nationAlt_tag = movie.find('nationAlt')# 제작국가
    nationAlt = nationAlt_tag.text if nationAlt_tag is not None else None

    genreAlt_tag = movie.find('genreAlt')# 장르
    genreAlt = genreAlt_tag.text if genreAlt_tag is not None else None

    peopleNm_tag = movie.find('peopleNm')# 감독
    peopleNm = peopleNm_tag.text if peopleNm_tag is not None else None

    movies_list.append({'movieCd': movieCd, 'movieNm': movieNm, 'movieNmEn': movieNmEn,
                        'openDt': openDt, 'nationAlt': nationAlt, 'genreAlt': genreAlt, 'peopleNm': peopleNm})
```
